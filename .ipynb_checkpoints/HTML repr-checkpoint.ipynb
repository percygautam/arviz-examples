```python
# Generating HTML repr
import arviz as az
data = az.load_arviz_data('centered_eight')
import xarray
xarray.set_options(display_style="html")
data.posterior
```

```python
html_template = """
<div>
  <div class='xr-header'>
    <div class="xr-dataarray-cls">arviz.InferenceData</div>
    <div class="xr-dataarray-name">'centered-eight'</div>
  </div>
  {}
</div>
"""
element_temp ='''
<ul class="xr-sections">
<li>
      <input id="xr-dataset-coord_{0}" type="checkbox">
      <label for="xr-dataset-coord_{0}">{0}</label>
<ul id="xr-dataset-coord-list" class="xr-var-list">
<div>{1}</div>
</ul>
</li>
</ul>
'''
css_template = """
<style>
body {
  /* jupyterlab settings */
  font-family: "Helvetica Neue", Helvetica, Arial, sans-serif;
  font-size: 13px;
  line-height: 1.5;
  background-color: #fff;
}

.xr-wrap {
  width: 700px;
}

.xr-header {
  padding: 6px 0 6px 3px;
  border-bottom: solid 1px #777;
  color: #555;;
}

.xr-header > div,
.xr-header > ul {
  display: inline;
  margin-top: 0;
  margin-bottom: 0;
}

.xr-dataarray-cls,
.xr-dataarray-name {
  margin-left: 2px;
  margin-right: 10px;
}

.xr-dataarray-name {
  color: #000;
}

.xr-sections {
  list-style: none;
  padding: 3px;
  margin: 0;
}

.xr-sections input {
  display: none;
}

.xr-sections label {
  display: inline-block;
}

.xr-sections input:enabled + label {
  cursor: pointer;
}

.xr-sections input:not(.xr-values-in) ~ ul {
  display: none;
}

.xr-sections input:not(.xr-values-in):checked ~ ul {
  display: block;
}

.xr-sections > li > input + label {
  width: 300px;
  color: #555;
  font-weight: 500;
  padding: 4px 0 2px 0;
}

.xr-sections > li > input + label:before {
  display: inline-block;
  content: '►';
  font-size: 11px;
  width: 15px;
  text-align: center;
}

.xr-sections > li > input:disabled + label:before {
  color: #777;
}

.xr-sections > li > input:checked + label:before {
  content: '▼';
}
</style>
"""
```

```python
template = ""
elements=""
for data_group in data._groups:
    func = getattr(data,data_group)
    elements += element_temp.format(data_group, func._repr_html_())
template += html_template.format(elements) 
template += css_template
```

```python
class HTML_example:
    
    def _repr_html_(self):
        return template

ex = HTML_example()
ex
```

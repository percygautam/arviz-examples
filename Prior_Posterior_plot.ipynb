```python
# Front end
import numpy as np
import arviz as az
import matplotlib.pyplot as plt

from arviz.plots.plot_utils import (
    xarray_var_iter,
    _scale_fig_size,
    default_grid,
    filter_plotters_list,
    get_plotting_function,
)

def plotter_gen(prior, prior_vars, coords):
    for key in coords.keys():
        coords[key] = np.where(np.in1d(prior[key], coords[key]))[0]
        
    obs_plotters = filter_plotters_list(
        list(
            xarray_var_iter(
                prior.isel(coords), var_names=prior_vars, combined=True
            )
        ),
        "plot_ppp",
    )
    length_plotters = len(obs_plotters)
    plotters = [
        tup
        for _, tup in zip(
            range(length_plotters),
            xarray_var_iter(
                prior.isel(coords),
                var_names=prior_vars,
#                skip_dims=set(flatten_pp),
                combined=True,
            ),
        )
    ]
    return plotters
```

```python
def plot_pp(data, prior_vars=None, posterior_vars=None, observed_vars=None, prior_coords=None, 
                  posterior_coords=None, observed_coords={}):
    
    prior_def = data.prior.defs.sel(team="Italy")
    post_def = data.posterior.defs.sel(team="Italy")
    observed_def = data.observed_data.home_points
    prior = data.prior

    #DEFINE PLOTTERS
    pp_prior_plotters = plotter_gen(data.prior, prior_vars=prior_vars, coords= prior_coords)
    pp_posterior_plotters = plotter_gen(data.posterior, prior_vars=posterior_vars, coords= posterior_coords)
    pp_ob_plotters = plotter_gen(data.observed_data, prior_vars=observed_vars, coords= observed_coords)
    
    
    # ORIGINAL GRAPHS
    fig = plt.figure(figsize=(10, 7))
    ax1 = plt.subplot2grid((2, 3), (0, 0), fig=fig)
    ax2 = plt.subplot2grid((2, 3), (0, 1), fig=fig)
    ax3 = plt.subplot2grid((2, 3), (0, 2), fig=fig)
    ax4 = plt.subplot2grid((2, 3), (1, 0), colspan=3, fig=fig)

    az.plot_kde(prior_def, ax=ax1)
    ax1.set_title("Prior")

    az.plot_kde(post_def, plot_kwargs={'color':'g'}, ax=ax2)
    ax2.set_title("Posterior")
    
    az.plot_kde(observed_def, plot_kwargs={'color':'r'},ax=ax3)
    ax3.set_title("Observed")

    az.plot_kde(prior_def, ax=ax4)
    az.plot_kde(post_def.values, plot_kwargs={'color':'g'}, ax=ax4)
    az.plot_kde(observed_def, plot_kwargs={'color':'r'}, ax=ax4)
    ax4.set_title("Both");
    
    
    #GRAPHS BY PLOTTERS
    fig = plt.figure(figsize=(10, 7))
    plt.title("By plotters")
    ax1 = plt.subplot2grid((2, 3), (0, 0), fig=fig)
    ax2 = plt.subplot2grid((2, 3), (0, 1), fig=fig)
    ax3 = plt.subplot2grid((2, 3), (0, 2), fig=fig)
    ax4 = plt.subplot2grid((2, 3), (1, 0), colspan=3, fig=fig)

    az.plot_kde(pp_prior_plotters[0][2], ax=ax1)
    ax1.set_title("Prior")

    az.plot_kde(pp_posterior_plotters[0][2], plot_kwargs={'color':'g'}, ax=ax2)
    ax2.set_title("Posterior")
    
    
    li = []
    for i in pp_ob_plotters:
        li.append(i[2])
    az.plot_kde(np.array(li).flatten(), plot_kwargs={'color':'r'},ax=ax3)
    ax3.set_title("Observed")


    az.plot_kde(pp_prior_plotters[0][2], ax=ax4)
    az.plot_kde(pp_posterior_plotters[0][2], plot_kwargs={'color':'g'}, ax=ax4)
    az.plot_kde(np.array(li).flatten(), plot_kwargs={'color':'r'}, ax=ax4)
    ax4.set_title("Both");
    
    
```

```python
data = az.load_arviz_data("rugby")
plot_pp(data, prior_vars=["defs"], posterior_vars=["defs"],
              prior_coords={"team":"Italy"},
              posterior_coords={"team":"Italy"}, observed_vars=["home_points"],)
```

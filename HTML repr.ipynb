```python
# Generating HTML repr
import arviz as az
data = az.load_arviz_data('centered_eight')
data.posterior
```

```python
import xarray
xarray.set_options(display_style="html")
data.posterior
```

```python
html_template = """
<div>
  <div class='xr-header'>
    <div class="xr-obj-type">arviz.InferenceData</div>
  </div>
  {}
</div>
"""
element_temp ='''
<ul class="xr-sections">
<li class = "xr-section-item">
      <input id="xr-dataset-coord_{0}" class="xr-section-summary-in" type="checkbox">
      <label for="xr-dataset-coord_{0}" class = "xr-section-summary">{0}</label>
      <div class="xr-section-inline-details"></div>
      <div class="xr-section-details">
          <ul id="xr-dataset-coord-list" class="xr-var-list">
              <div>{1}</div>
          </ul>
      </div>
</li>
</ul>
'''
css_template = "<style>" + xarray.core.formatting_html.CSS_STYLE + ".xr-wrap{width:700px!important;}" + "</style>"
```

```python
template = ""
elements=""
for data_group in data._groups:
    func = getattr(data,data_group)
    elements += element_temp.format(data_group, func._repr_html_())
template += html_template.format(elements) 
template += css_template
```

```python
class HTML_example:
    
    def _repr_html_(self):
        return template

ex = HTML_example()
ex
```
